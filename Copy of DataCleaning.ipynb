```python
phase2_df = df_clean.copy()

print("Phase 2 copy created")
print("Shape:", phase2_df.shape)
```

```python
important_columns = [
    'date',
    'monthly_income',
    'monthly_expense_total',
    'budget_goal',
    'savings_rate',
    'actual_savings',
    'transaction_count',
    'savings_goal_met'
]

display(phase2_df[important_columns].head(10))
```

```python
check_df = phase2_df[
    [
        'monthly_income',
        'monthly_expense_total',
        'budget_goal',
        'savings_rate',
        'actual_savings',
        'savings_goal_met'
    ]
].copy()

check_df['calculated_savings'] = (
    check_df['monthly_income']
    - check_df['monthly_expense_total']
)

check_df['actual_savings_difference'] = (
    check_df['actual_savings']
    - check_df['calculated_savings']
).abs()

display(check_df.head(10))

print(
    "Average difference between actual_savings and "
    "income minus expense:",
    check_df['actual_savings_difference'].mean()
)
```

```python
rate_check = phase2_df[
    [
        'monthly_income',
        'actual_savings',
        'savings_rate'
    ]
].copy()

rate_check['calculated_decimal_rate'] = (
    rate_check['actual_savings']
    / rate_check['monthly_income']
)

rate_check['calculated_percentage_rate'] = (
    rate_check['actual_savings']
    / rate_check['monthly_income']
) * 100

decimal_difference = (
    rate_check['savings_rate']
    - rate_check['calculated_decimal_rate']
).abs().mean()

percentage_difference = (
    rate_check['savings_rate']
    - rate_check['calculated_percentage_rate']
).abs().mean()

print("Savings rate minimum:",
      phase2_df['savings_rate'].min())

print("Savings rate maximum:",
      phase2_df['savings_rate'].max())

print("Decimal-rate average difference:",
      decimal_difference)

print("Percentage-rate average difference:",
      percentage_difference)
```

```python
goal_check = phase2_df[
    [
        'monthly_income',
        'monthly_expense_total',
        'budget_goal',
        'actual_savings',
        'savings_goal_met'
    ]
].copy()

# Possibility 1:
# budget_goal is a savings target
goal_check['result_if_savings_target'] = (
    goal_check['actual_savings']
    >= goal_check['budget_goal']
).astype(int)

# Possibility 2:
# budget_goal is an expense limit
goal_check['result_if_expense_limit'] = (
    goal_check['monthly_expense_total']
    <= goal_check['budget_goal']
).astype(int)

savings_target_agreement = (
    goal_check['result_if_savings_target']
    == goal_check['savings_goal_met']
).mean() * 100

expense_limit_agreement = (
    goal_check['result_if_expense_limit']
    == goal_check['savings_goal_met']
).mean() * 100

print(
    "Agreement if budget_goal is savings_target:",
    round(savings_target_agreement, 2),
    "%"
)

print(
    "Agreement if budget_goal is expense_limit:",
    round(expense_limit_agreement, 2),
    "%"
)

display(goal_check.head(15))
```

```python
final_df = pd.DataFrame()

print("Empty final dataset created")
```

```python
# Existing date will be treated as the prediction date
final_df['prediction_date'] = pd.to_datetime(
    phase2_df['date'],
    errors='coerce'
)

# Extract year
final_df['year'] = final_df['prediction_date'].dt.year

# Extract month name
final_df['month'] = final_df['prediction_date'].dt.month_name()
```

```python
final_df['income_received_date'] = (
    final_df['prediction_date']
    .dt.to_period('M')
    .dt.to_timestamp()
)
```

```python
final_df['income_amount'] = phase2_df['monthly_income']

final_df['savings_target'] = phase2_df['budget_goal']

final_df['amount_spent_so_far'] = (
    phase2_df['monthly_expense_total']
)

final_df['transaction_count'] = (
    phase2_df['transaction_count']
)

final_df['target_achieved'] = (
    phase2_df['savings_goal_met'].astype(int)
)
```

```python
final_df['expense_limit'] = (
    final_df['income_amount']
    - final_df['savings_target']
)
```

```python
final_df['remaining_budget'] = (
    final_df['expense_limit']
    - final_df['amount_spent_so_far']
)
```

```python
final_df['budget_used_percentage'] = np.where(
    final_df['expense_limit'] > 0,
    (
        final_df['amount_spent_so_far']
        / final_df['expense_limit']
    ) * 100,
    np.nan
)
```

```python
final_df['budget_used_percentage'] = (
    final_df['budget_used_percentage'].round(2)
)
```

```python
days_in_month = (
    final_df['prediction_date'].dt.days_in_month
)

current_day = (
    final_df['prediction_date'].dt.day
)

final_df['days_remaining'] = (
    days_in_month - current_day
)
```

```python
final_columns = [
    'year',
    'month',
    'income_received_date',
    'prediction_date',
    'income_amount',
    'expense_limit',
    'savings_target',
    'amount_spent_so_far',
    'transaction_count',
    'remaining_budget',
    'budget_used_percentage',
    'days_remaining',
    'target_achieved'
]

final_df = final_df[final_columns]
```

```python
invalid_plan = final_df[
    (
        final_df['expense_limit']
        + final_df['savings_target']
    )
    > final_df['income_amount']
]

print("Invalid financial plans:", len(invalid_plan))
```

```python
# 1. Check the original difference
final_df['plan_difference'] = (
    final_df['expense_limit']
    + final_df['savings_target']
    - final_df['income_amount']
)

display(
    final_df[
        final_df['plan_difference'] > 0
    ][[
        'income_amount',
        'expense_limit',
        'savings_target',
        'plan_difference'
    ]]
)
```

```python
# 2. Round all money columns
money_columns = [
    'income_amount',
    'expense_limit',
    'savings_target',
    'amount_spent_so_far',
    'remaining_budget'
]

final_df[money_columns] = final_df[money_columns].round(2)
```

```python
# 3. Recalculate the difference after rounding
final_df['plan_difference'] = (
    final_df['expense_limit']
    + final_df['savings_target']
    - final_df['income_amount']
).round(2)
```

```python
# 4. Validate using one-paisa tolerance
tolerance = 0.01

invalid_plan = final_df[
    final_df['plan_difference'] > tolerance
]

print("Truly invalid financial plans:", len(invalid_plan))
```

```python
# 5. Perform additional checks
print(
    "Savings target greater than income:",
    (
        final_df['savings_target']
        > final_df['income_amount']
    ).sum()
)

print(
    "Zero or negative expense limits:",
    (
        final_df['expense_limit'] <= 0
    ).sum()
)
```

```python
# 6. Remove the temporary checking column
final_df.drop(
    columns='plan_difference',
    inplace=True
)
```

```python
money_columns = [
    'income_amount',
    'expense_limit',
    'savings_target',
    'amount_spent_so_far',
    'remaining_budget'
]

final_df[money_columns] = final_df[money_columns].round(2)
```
